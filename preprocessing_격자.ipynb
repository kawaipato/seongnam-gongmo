```python
cctt = pd.DataFrame({'0' : [ct[ct.index == 0][['cctv']].sum()[0]], 'Other' : [ct[ct.index >=1][['cctv']].sum()[0]]})
cctt
```

```python
plt.bar(cctt.columns,cctt.loc[0])
plt.show()
```

```python
plt.figure(figsize=(15,10))
sns.barplot(data = light, x = 'index', y = '보안등')
# plt.xticks(rotation=45)
plt.show()
```

```python
lightt = pd.DataFrame({'0' : [light[light.index == 0][['보안등']].sum()[0]], 'Other' : [light[light.index >=1][['보안등']].sum()[0]]})
lightt
```

```python
plt.bar(lightt.columns,lightt.loc[0])
plt.show()
```

```python
df2 = df.copy()
df2['security']=df2.cctv + df2.보안등
df2.head()
```

```python
s = df2['security'].value_counts().reset_index()
print('light 0\'s : {0}, cctv 0\'s : {1}, light+cctv 0\
s : {2}'.format(light.iloc[0][1],ct.iloc[0][1],s.iloc[0][1]))
s
```

```python
all_s = pd.DataFrame({'0' : [s[s.index == 0][['security']].sum()[0]], 'Other' : [s[s.index >=1][['security']].sum()[0]]})
all_s
```

```python
plt.bar(all_s.columns,all_s.loc[0])
plt.show()
```

```python
def transform_label(label):
    if label == 0:
        return 0
    else:
        return 1

# 열 변환하기
df2['security_encoded'] = df2['security'].map(transform_label)
```
